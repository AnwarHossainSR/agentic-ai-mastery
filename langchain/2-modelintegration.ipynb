### Model Integration with GEMINI, OPENAI and GROQ

```python
import os
from dotenv import load_dotenv
load_dotenv()

# Set all the environment variables
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("google_genai:gemini-2.5-flash-lite")
response = model.invoke([("user", "Why is the sky blue?")])
print(response.text)


```

### Stream response

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("google_genai:gemini-2.5-flash-lite")

for chunk in model.stream([("user", "Write me 1000 words about the history of the Eiffel Tower.")]):
    print(chunk.text, end="")
```

### Batch

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("google_genai:gemini-2.5-flash-lite")

response = model.batch(
    [
        [("user", "What is the capital of France?")],
        [("user", "What is the tallest mountain in the world?")],
        [("user", "Who wrote 'Pride and Prejudice'?")]
    ]
)

for res in response:
    print(res.text)
```
