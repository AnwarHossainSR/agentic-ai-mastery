### Langchain Version: 1.1.0

```python
import os
from dotenv import load_dotenv
load_dotenv()

# Set all the environment variables
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from typing import Optional
from langchain.agents import create_agent
from langchain_core.tools import tool

@tool
def get_weather(city: Optional[str] = None) -> str:
    "Get the current weather for a given location."
    city = city or "New York"
    return f"The current weather in {city} is sunny with a temperature of 25°C."

# Force API-key based Gemini route (not Vertex AI project auth)
os.environ["GOOGLE_GENAI_USE_VERTEXAI"] = "False"

agent = create_agent(
    model="google_genai:gemini-2.5-flash",
    tools=[get_weather],
    system_prompt="You are a helpful assistant that can use tools to answer questions and perform tasks.",
)
agent.invoke({"messages": [("user", "What is the current weather in New Yark?")]})
```
